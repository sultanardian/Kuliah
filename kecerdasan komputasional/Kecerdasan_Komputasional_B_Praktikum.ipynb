## Missing Value
Periksa apakah terdapat data yang tidak memiliki nilai.
Tentukan penanganan seperti apa yang sebaiknya digunakan serta jelaskan kenapa menggunakan metode tersebut! 


````python
df = pd.read_csv('https://raw.githubusercontent.com/tiodh/nilai/main/nilai.csv')
df
````

````python
import numpy as np

for index, row in df.iterrows():
    if row['tugas'] == 'None':
        row['tugas'] = np.nan
    row['tugas'] = float(row['tugas'])

    if row['uts'] == 'None':
        row['uts'] = np.nan
    row['uts'] = float(row['uts'])

    if row['uas'] == 'None':
        row['uas'] = np.nan
    row['uas'] = float(row['uas'])

df
````

````python
df.isnull().sum()
````

````python
df.fillna(df.mean()).std()
````

````python
df.fillna(df.mean(), inplace = True)
df.isnull().sum()
````

Diambil nilai rata rata atau ``` mean() ``` sebagai data missing value karena menghasilkan nilai standart deviation terkecil

## Outlier Detection
Cobalah melakuan pendeteksian outlier dengan menggunakan Interquartile Range (IQR). Kemudian tentukan metode penanganan dan jelaskan kenapa menggunakan metode penanganan tersebut!

````python
Q1 = df.quantile(.25)
Q3 = df.quantile(.75)

IQR = Q3 - Q1

non_outlier = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR)))]
non_outlier
````

````python
clean_outlier = non_outlier.fillna(non_outlier.min())
clean_outlier
````

Diambil nilai minimal atau ```min() ``` sebagai pengganti data outlier karena tidak menghasilkan outlier baru
