```python
import pandas as pd
```

```python
co2 = pd.read_csv('https://raw.githubusercontent.com/TrainingByPackt/Interactive-Data-Visualization-with-Python/master/datasets/co2.csv')
co2.head()
```

```python
gm = pd.read_csv('https://raw.githubusercontent.com/TrainingByPackt/Interactive-Data-Visualization-with-Python/master/datasets/gapminder.csv')
gm.rename(columns = {'Country' : 'country gm', 'Year' : 'year'}, inplace = True)
gm.head()
```

```python
df_gm = gm[['country gm', 'region']].drop_duplicates()
```

```python
df_w_regions = pd.merge(co2, df_gm, left_on = 'country', right_on = 'country gm', how = 'inner')
df_w_regions.drop('country gm', axis = 1, inplace = True)
df_w_regions
```

```python
melted_co2_df = pd.melt(df_w_regions, id_vars = ['country', 'region']).sort_values(by = ['country', 'variable'], axis = 0)
melted_co2_df.rename(columns = {'variable' : 'year', 'value' : 'co2'}, inplace = True)
melted_co2_df.dropna(inplace = True)
melted_co2_df.reset_index(drop = True, inplace = True)
melted_co2_df['year'] = melted_co2_df['year'].astype('int64')
melted_co2_df
```

```python
df_gdp = gm[['gdp', 'year', 'country gm']]
data = pd.merge(melted_co2_df, df_gdp, left_on = ['country', 'year'], right_on = ['country gm', 'year'], how = 'right').dropna().reset_index(drop = True)
data.drop('country gm', axis = 1, inplace = True)
data
```

```python
import plotly.express as px
```

```python
fig = px.scatter(data,
                 x = 'year',
                 y = 'co2',
                 hover_name = 'country',
                 color = 'region',
                 width = 1579,
                 height = 400,
                 marginal_y = 'box')
fig.show()
```

```python
fig = px.scatter(data,
                 x = 'gdp',
                 y = 'co2',
                 hover_name = 'country',
                 color = 'region',
                 width = 1579,
                 height = 400,
                 marginal_x = 'rug',
                 marginal_y = 'box',
                 animation_frame = 'year',
                 animation_group = 'country',
                 log_x = True)
fig.show()
```

```python
fig = px.scatter(data, x = 'gdp', y = 'co2', color = 'country')
fig.show()
```

```python
fig = px.density_contour(data.head(1000), x = 'gdp', y = 'co2', color = 'region') 
fig.show()
```
